# Numeros Pseudoaleatorios

```python
import pandas as pd

x0 = int(input("Ingrese la semilla: "))
iteraciones=int(input("Ingrese el número de iteraciones: "))
digitos=int(input("Ingrese el número de digitos: "))
iteracion=0
Xn=x0
Xnx=[]
xn2=[]
i=[]
long=[]
ui=[]
rn=[]
while(iteracion<=iteraciones):
    Xc=int(Xn) * int(Xn)
    Xcs=str(Xc)
    while(len(Xcs) < 8):
        Xcs="0"+Xcs

    Longitud=len(Xcs)
    Ui=Xcs[int((Longitud/2)-2):int((Longitud/2)+2)]
    Nn="1"
    while(len(Nn)<(digitos+1)):
        Nn=Nn+"0"
    Rn=int(Ui)/int(Nn)    
    #print("Iteración: ",iteracion)
    i.append(iteracion)
    #print("Xn: ",Xn)
    Xnx.append(Xn)
    #print("Xn * Xn: ",Xcs)
    xn2.append(Xcs)
    #print("Longitud: ",Longitud)
    long.append(Longitud)
    #print("Ui: ",Ui)
    ui.append(Ui)
    #print(Nn)
    #print("Randomico: ",Rn)
    rn.append(Rn)
    Xn=Ui
    iteracion=iteracion+1
    
data={'Iteracion':i,'Xn':Xnx,'Xn*Xn':xn2,'Longitud':long,'Ui':ui,'Rn':rn}
df=pd.DataFrame(data)
print("Digitos: ",digitos)
df
```
