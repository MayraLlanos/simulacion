# Congruencia Lineal

```python
import pandas as pd
a=int(input("Ingrese \"a\": "))
b=int(input("Ingrese \"c\": "))
m=int(input("Ingrese \"m\": "))
x0=int(input("Ingrese la semilla: "))
iteraciones=int(input("Ingrese número de iteraciones: "))
Unn=0
Xn=[]
Un=[]
iteracioness=[]
Xnn=x0
i=0
iteracioness.append(i)
i=i+1
Xn.append(Xnn)
Un.append(Unn)
while(i <= iteraciones):
    iteracioness.append(i)
    Xnn= (a * Xnn + b) % m
    Xn.append(Xnn)
    Unn=Xnn/m
    Un.append(Unn)
    i=i+1
data={'iteracion':iteracioness,'Xn':Xn,'Un':Un}
df=pd.DataFrame(data)
df
```
